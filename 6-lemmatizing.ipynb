```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lem = WordNetLemmatizer()
```

```python
words = ['drive', 'driving', 'driver', 'drives', 'drove', 'cats', 'children']
```

```python
for w in words:
    print(lem.lemmatize(w))
```

```python
lem.lemmatize('drove', 'v')
```
